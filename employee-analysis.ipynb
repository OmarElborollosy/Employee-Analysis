```python
# for basic operations
import numpy as np
import pandas as pd

# for data visualizations
import seaborn as sns
import matplotlib.pyplot as plt

import scipy.stats as stats
import math

import statsmodels.api as sm

```

```python
# reading the data
data = pd.read_csv('/kaggle/input/employee/Employee.csv')

# lets check the shape of the dataset
data.shape
```

```python
data.columns
```

What is the average age of employees in the company?

```python
data['Age'].value_counts()
```

```python
sample_ages = np.random.choice(a= data['Age'], size=1000)

print ("Sample mean:", sample_ages.mean() )
```

```python

# lets seed the random values
np.random.seed(10)

# lets take a sample size
sample_size = 500
sample = np.random.choice(a= data['Age'],
                          size = sample_size)
sample_mean = sample.mean()

# Get the z-critical value*
z_critical = stats.norm.ppf(q = 0.95)

 # Check the z-critical value
print("z-critical value: ",z_critical)

# Get the population standard deviation
pop_stdev = data['Age'].std()

# checking the margin of error
margin_of_error = z_critical * (pop_stdev/math.sqrt(sample_size))

# defining our confidence interval
confidence_interval = (sample_mean - margin_of_error,
                       sample_mean + margin_of_error)

# lets print the results
print("Confidence interval:",end=" ")
print(confidence_interval)
print("True mean: {}".format(data['Age'].mean()))
```

We can conclude that the average age of the employees in this company is 29y and the percentage of interns and old people aging 40y or more is very small.

What is the percentage of female employees in the company comparing it to male percentage?

```python
data['Gender'] = data['Gender'].replace({0: 'Female', 1: 'Male'})

print(data['Gender'].value_counts())
n = data.shape[0]
Female = data['Gender'].value_counts().loc['Female']

print("\nTotal Observation ==>",n,"\t","Number of people of female employees in our company is==> ",Female,"\n")
```

```python
sample_Size = 500
sample_gender = np.random.choice(a= data['Gender'],
                          size = sample_Size)


n_sample = sample_gender.shape[0]
Female_sample = (sample_gender=='Female').sum()

print("\nTotal Observation ==>",n_sample,"\t","Number of female employees in our sample data is ==> ",Female_sample,"\n")
```

```python
print("\n95% Confidence interval with statsmodels library ==>",sm.stats.proportion_confint(Female_sample, n_sample),"\n")
```

We can conclude that the percentage of female employees is between 35% to 43% so the male employees took the bigger priority

```python
print("\n95% Confidence interval with statsmodels library ==>",sm.stats.proportion_confint(Bachelors_sample, n_sample),"\n")
```
